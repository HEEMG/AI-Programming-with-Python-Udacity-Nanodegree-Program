While Loops

For loops are an example of "definite iteration" meaning that the loop's body is run a predefined number of times. This differs from "indefinite iteration" which is when a loop repeats an unknown number of times and ends when some condition is met, which is what happens in a while loop. Here's an example of a while loop.

```python
card_deck = [4, 11, 8, 5, 13, 2, 8, 10]
hand = []

## adds the last element of the card_deck list to the hand list
## until the values in hand add up to 17 or more
while sum(hand)  < 18:
    hand.append(card_deck.pop())
```

### Explanation

1.This code appears to simulate a simple card game where you have a deck of cards represented by the list card_deck and you want to draw cards one by one and add them to your hand (represented by the list hand) until the sum of the values in your hand is 17 or more. Let's break down the code step by step:

2.card_deck is a list that contains the following integers: [4, 11, 8, 5, 13, 2, 8, 10]. These integers presumably represent the values of cards in a card game.

3.hand is an empty list that will be used to store the cards you draw from the deck.

4.The code enters a while loop with the condition while sum(hand) < 18:. This means the loop will continue to run as long as the sum of the values in your hand is less than 18.

5.Inside the loop, card_deck.pop() is used to remove and return the last element from the card_deck list, which simulates drawing a card from the deck. The popped card is then appended to the hand list using hand.append(card_deck.pop()).

6.The loop continues to draw and add cards to your hand until the sum of the values in your hand becomes 17 or more.

7.Once the sum of your hand reaches 17 or more, the loop terminates, and the code prints the contents of your hand using print(hand).

In summary, this code keeps drawing cards from the deck and adding them to your hand until the sum of the card values in your hand is 17 or greater. It then prints the cards in your hand at the end. The specific cards you end up with will depend on the values in the card_deck, and the order in which they are drawn.

# Break, Continue

Break, Continue
Sometimes we need more control over when a loop should end, or skip an iteration. In these cases, we use the break and continue keywords, which can be used in both for and while loops.

break terminates a loop
continue skips one iteration of a loop

```python
manifest = [("bananas", 15), ("mattresses", 24), ("dog kennels", 42), ("machine", 120), ("cheeses", 5)]

# the code breaks the loop when weight exceeds or reaches the limit
print("METHOD 1")
weight = 0
items = []
for cargo_name, cargo_weight in manifest:
    print("current weight: {}".format(weight))
    if weight >= 100:
        print("  breaking loop now!")
        break
    else:
        print("  adding {} ({})".format(cargo_name, cargo_weight))
        items.append(cargo_name)
        weight += cargo_weight

print("\nFinal Weight: {}".format(weight))
print("Final Items: {}".format(items))

# skips an iteration when adding an item would exceed the limit
# breaks the loop if weight is exactly the value of the limit
print("\nMETHOD 2")
weight = 0
items = []
for cargo_name, cargo_weight in manifest:
    print("current weight: {}".format(weight))
    if weight >= 100:
        print("  breaking from the loop now!")
        break
    elif weight + cargo_weight > 100:
        print("  skipping {} ({})".format(cargo_name, cargo_weight))
        continue
    else:
        print("  adding {} ({})".format(cargo_name, cargo_weight))
        items.append(cargo_name)
        weight += cargo_weight

print("\nFinal Weight: {}".format(weight))
print("Final Items: {}".format(items))
```

Zip and Enumerate

zip and enumerate are useful built-in functions that can come in handy when dealing with loops.

Zip
zip returns an iterator that combines multiple iterables into one sequence of tuples. Each tuple contains the elements in that position from all the iterables. For example, printing

list(zip(['a', 'b', 'c'], [1, 2, 3])) would output [('a', 1), ('b', 2), ('c', 3)].

Like we did for range() we need to convert it to a list or iterate through it with a loop to see the elements.

You could unpack each tuple in a for loop like this.

```python
letters = ['a', 'b', 'c']
nums = [1, 2, 3]

for letter, num in zip(letters, nums):
    print("{}: {}".format(letter, num))
```

```python
rivers = ['Jordan', 'Nile', 'Victoria', 'Amazon']

continents = ['Asia','Africa', 'Africa', 'Africa']

for river,continent in zip(rivers, continents):
    print("{}:{}".format(river,continent))
```
